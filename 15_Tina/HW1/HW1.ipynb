```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 參數設定
p = 0.3
n = 50
n_experiments = 10

# 執行模擬
sample_means = [np.random.binomial(1, p, n).mean() for _ in range(n_experiments)]

# 計算理論上的平均值與標準差
mu = p
sigma = np.sqrt(p * (1 - p) / n)

# 繪圖
plt.hist(sample_means, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label='Simulated')

# 繪製理論常態分佈曲線
x = np.linspace(0, 1, 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2, label='Theoretical (Normal)')

plt.title(f'CLT: Bernoulli Means (n={n}, p={p})')
plt.xlabel('Sample Mean Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
